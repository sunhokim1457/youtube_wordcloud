```python
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError

DEVELOPER_KEY = 'AIzaSyAoMzPjwS1LxBs1XtrWxq-eRiC-ot4Gcr8'
YOUTUBE_API_SERVICE_NAME = 'youtube'
YOUTUBE_API_VERSION ='v3'

youtube=build(YOUTUBE_API_SERVICE_NAME,YOUTUBE_API_VERSION,developerKey=DEVELOPER_KEY)

# ìˆœì„œëŒ€ë¡œ í”¼ì§€ì»¬ ê°¤ëŸ¬ë¦¬, í•ë¸”ë¦¬, ì§€í”¼í‹°
channelId_list = ['UCdtRAcd3L_UpV4tMXCw63NQ', 'UC3hRpIQ4x5niJDwjajQSVPg', 'UCoHirUZkTs1nOtyJTNbNJ5g']
channelRecentList = []

for id in channelId_list:
    recent_playlist = youtube.search().list(
    part = 'snippet',
    channelId = id,
    maxResults = 30,
    order = 'date',
    publishedAfter = '2023-03-01T00:00:00Z'
    ).execute()
    
    channelRecentList.append(recent_playlist)
    
channelRecentList[0]
```

```python
# ê° ì¬ìƒëª©ë¡ì˜ í†µê³„ ê³„ì‚°
channel_viewCount = [] # ì´ ì¡°íšŒìˆ˜
channel_commentCount = [] # ì´ ë§ê¸€ìˆ˜
channel_likeCount = [] # ì´ ì¢‹ì•„ìš” ê°œìˆ˜
channel_videoCount = [] # ì´ ë¹„ë””ì˜¤ ê°œìˆ˜

# ê° ì¬ìƒëª©ë¡ë§ˆë‹¤ ë°˜ë³µ

for i in range(len(channelRecentList)):
    channel_viewCount.append(0)
    channel_commentCount.append(0)
    channel_likeCount.append(0)
    channel_videoCount.append(0) 

    # ì¬ìƒëª©ë¡ì— ë“¤ì–´ìˆëŠ” ë¹„ë””ì˜¤ ì •ë³´ë¥¼ ì‚¬ìš©í•˜ì—¬ ê³„ì‚°
    for video in channelRecentList[i]['items']:
        try:   # ì–´ë–¤ ë¹„ë””ì˜¤ì˜ videoIdê°€ ì¡´ì¬í•˜ì§€ ì•Šì•„ ì„ì‹œ ì˜ˆì™¸ì²˜ë¦¬   
            videoStat = youtube.videos().list(
                part = 'snippet ,statistics',
                id = video['id']['videoId']
            ).execute()
            
            if videoStat['items']: # ë¹„ê³µê°œ ë™ì˜ìƒì€ ì œì™¸
                channel_viewCount[-1] += int(videoStat['items'][0]['statistics']['viewCount'])
                channel_commentCount[-1] += int(videoStat['items'][0]['statistics']['commentCount'])
                channel_likeCount[-1] += int(videoStat['items'][0]['statistics']['likeCount'])
                channel_videoCount[-1] += 1
        except:
            pass
            
channel_viewCount
```

```python
subscriberCount = []

for id in channelId_list:
    channel_response = youtube.channels().list(
        part="statistics",
        id=id
    ).execute()
    
    subscriberCount.append(channel_response['items'][0]['statistics']['subscriberCount'])
    
subscriberCount
```

```python
# DataFrame ë§Œë“¤ê¸°
# ì´ ë°ì´í„°
from pandas import DataFrame
count_total_data ={
    'subscriberCount' : subscriberCount,
    'viewCount' : channel_viewCount,
    'commentCount' : channel_commentCount,
    'likeCount' : channel_likeCount,
}
channel_title = ['í”¼ì§€ì»¬ ê°¤ëŸ¬ë¦¬', 'í•ë¸”ë¦¬', 'ì§€í”¼í‹°']
channelTable_total= DataFrame(count_total_data, index= channel_title)

channelTable_total
```

```python
# ê° ë°ì´í„° í‰ê· í™”
# í‰ê·  ë°ì´í„°
count_avg_data ={
    'subscriberCount' : subscriberCount,
    'viewCount' : [round(x/y) for x,y in zip(channel_viewCount, channel_videoCount)],
    'commentCount' : [round(x/y) for x,y in zip(channel_commentCount, channel_videoCount)],
    'likeCount' : [round(x/y) for x,y in zip(channel_likeCount, channel_videoCount)],
}

channelTable_avg = DataFrame(count_avg_data, index=channel_title)

channelTable_avg
```
